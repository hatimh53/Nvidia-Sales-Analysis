# **NVIDIA GPU SALES ANALYSIS 2026 📈**

```python
# Importing important libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Data 

df = pd.read_csv(r"nvidia_sales_2026.csv")
df.head()
```

```python
# Changing data-type:
df["sale_date"] = pd.to_datetime(df["sale_date"])

df.info()


# Finding null value:
df.isnull().sum()

df["bundle_addon"] = df["bundle_addon"].fillna("No addon")
```

## **Exploratory Data Analysis**

```python
# Revenue broken down by Region & Sales Channel:

pivot = df.pivot_table(values="revenue_usd", index="region", columns="sales_channel", aggfunc="sum")
pivot

# % of sales records from each regions:

df["region"].value_counts(normalize=True).round(2) * 100
```

### *From the above Pivot table, we can see the revenue by sales channel and regions. North America leads almost in all sales channel with a great revenue.Also North America has the highest sales contribution in all regions.*

### **📈 Yearly & Monthly Trends**

```python
# Sales month & year:

df["sales_year"] = df["sale_date"].dt.year
df["sales_month"] = df["sale_date"].dt.to_period("M")
```

```python
# Yearly sales & revenue:

yearly_sales = df.groupby("sales_year")["units_sold"].sum().reset_index()
yearly_revenue = df.groupby("sales_year")["revenue_usd"].sum().round().reset_index()


monthly_sales = df.groupby("sales_month")["units_sold"].sum().round().reset_index()
monthly_revenue = df.groupby("sales_month")["revenue_usd"].sum().round().reset_index()
Identify 

# Visual Analysis:

fig, axes = plt.subplots(1, 2, figsize=(10,5))

sns.barplot(data=yearly_revenue, x="sales_year", y="revenue_usd", ax=axes[0])
axes[0].set_title("Revenue by Year",fontfamily="Times New Roman",fontsize=15)
axes[0].set_xticks([0, 1, 2])
axes[0].set_xticklabels(["2024", "2025", "2026 (half-yearly)"])

sns.barplot(data=yearly_sales, x="sales_year", y="units_sold", ax=axes[1])
axes[1].set_title("Sales by Year",fontfamily="Times New Roman",fontsize=15)
axes[1].set_xticks([0, 1, 2])
axes[1].set_xticklabels(["2024", "2025", "2026 (half-yearly)"])

plt.tight_layout()
plt.show()
```

### *1.From the above analysis, we can se that the sales and revenue is consistantly increases over years.*
### *2.The sales and revenue numbers of 2026 nearly suprassed the numbers of year 2025 in the first 2 Quarters itelf.*

### **📊 GPU Analysis**

```python
# GPU Models sales & revenue:

model_sales = df.groupby("gpu_model")["units_sold"].sum().round().sort_values(ascending=False).reset_index()
model_revenue = df.groupby("gpu_model")["revenue_usd"].sum().round().sort_values(ascending=False).reset_index()

# GPU model avg satisfaction scores:

model_scores = df.groupby("gpu_model")["customer_satisfaction_score"].mean().round(2).sort_values(ascending=False).reset_index()

# GPU Family sales & revenue:

family_sales = df.groupby("gpu_family")["units_sold"].sum().round().sort_values(ascending=False).reset_index()
family_revenue = df.groupby("gpu_family")["revenue_usd"].sum().round().sort_values(ascending=False).reset_index()


# Visual Analysis:

fig, axes = plt.subplots(2,2, figsize=(12,10))

sns.barplot(data = model_sales, x ="gpu_model", y="units_sold", hue="gpu_model", palette="viridis", ax=axes[0,0])
axes[0,0].set_title("GPU Model Sales",fontfamily="Times New Roman",fontsize=15)
axes[0,0].tick_params(axis="x", rotation=90)

sns.barplot(data = model_revenue, x ="gpu_model", y="revenue_usd", hue="gpu_model", palette="viridis", ax=axes[0,1])
axes[0,1].set_title("GPU Model Revenue",fontfamily="Times New Roman",fontsize=15)
axes[0,1].tick_params(axis="x", rotation=90)

sns.barplot(data = family_sales, x ="gpu_family", y="units_sold",hue="gpu_family", palette="viridis", ax=axes[1,0])
axes[1,0].set_title("GPU Family Sales",fontfamily="Times New Roman",fontsize=15)

sns.barplot(data = family_revenue, x ="gpu_family", y="revenue_usd",hue="gpu_family", palette="viridis",ax=axes[1,1])
axes[1,1].set_title("GPU Family Revenue",fontfamily="Times New Roman",fontsize=15)


plt.tight_layout()
plt.show()
```

### *1.In the GPU Model Sales, RTX 5070 leads the charts. While in revenue, B200 GPU set leads all.
### *2.It is worth noting that the GPU set which have low sales, conrtibutes highest in revenue generation. For ex- B200 & H200 have low sales but leads in revenue generation.
### *3.In case of GPU Family, Consumer Gaming leads the sales charts. While Data Centre AI leads in revenue.
### *4.GPU Model "L40S" has the best customer satisfaction scores. Also the B200 Model has the second best score which also leads in revenue.

### *🛒 Channel & Segment Analysis*

```python
# Channel Sales & Revenue:

channel_sales = df.groupby("sales_channel")["units_sold"].sum().sort_values(ascending=False).reset_index()
channel_revenue = df.groupby("sales_channel")["revenue_usd"].sum().round().sort_values(ascending=False).reset_index()

# Segment Sales & Revenue:

segment_sales = df.groupby("customer_segment")["units_sold"].sum().sort_values(ascending=False).reset_index()
segment_revenue = df.groupby("customer_segment")["revenue_usd"].sum().round().sort_values(ascending=False).reset_index()

# Visual Analysis:

fig, axes = plt.subplots(2,2, figsize=(12,10))

sns.barplot(data = channel_sales, x ="sales_channel", y="units_sold", hue="sales_channel", palette="viridis", ax=axes[0,0])
axes[0,0].set_title("Channel Sales",fontfamily="Times New Roman",fontsize=15)
axes[0,0].tick_params(axis="x", rotation=45)

sns.barplot(data = channel_revenue, x ="sales_channel", y="revenue_usd", hue="sales_channel", palette="viridis", ax=axes[0,1])
axes[0,1].set_title("Channel Revenue",fontfamily="Times New Roman",fontsize=15)
axes[0,1].tick_params(axis="x", rotation=45)

sns.barplot(data = segment_sales, x ="customer_segment", y="units_sold",hue="customer_segment", palette="viridis", ax=axes[1,0])
axes[1,0].set_title("Segment Sales",fontfamily="Times New Roman",fontsize=15)
axes[1,0].tick_params(axis="x", rotation=45)

sns.barplot(data = segment_revenue, x ="customer_segment", y="revenue_usd",hue="customer_segment", palette="viridis",ax=axes[1,1])
axes[1,1].set_title("Segment Revenue",fontfamily="Times New Roman",fontsize=15)
axes[1,1].tick_params(axis="x", rotation=45)


plt.tight_layout()
plt.show()
```

### *1.Retail/Etail channel contributes the highest sales along with revenue.*
### *2.Gaming contributes the highest sales and revenue among all customer segments.*

```python
# Price Premium Pct by Stock Status

stock_premium = df.groupby("stock_status")["price_premium_pct"].mean().round(2).sort_values(ascending=False).reset_index()

# Visual Analysis: 

plt.figure(figsize=(7,5))
sns.boxplot(data=df, x="stock_status", y="price_premium_pct")
plt.title("Price Premium Distribution by Stock Status",fontfamily="Times New Roman",fontsize=15)
plt.xlabel("Stock Status")
plt.ylabel("Price Premium (%)")
plt.show()
```

### 💡 Stock Scarcity Drives Price Premiums:

### *1.Median price premium rises consistently as availability drops.*
### *2.Sold Out items show the widest spread in premiums — buyers pay anywhere from ~25% to over 90% above MSRP, suggesting scarcity leads to unpredictable, sometimes extreme pricing.*
### *3.Even Backordered items show several high outliers (55-70% premium), meaning some buyers pay steep markups well before an item is fully sold out.*

```python
# MSRP vs Street Price:

plt.figure(figsize=(7,6))
sns.scatterplot(data=df, x="msrp_usd", y="avg_street_price_usd", hue="stock_status")
plt.title("MSRP vs Actual Street Price",fontfamily="Times New Roman",fontsize=15)
plt.xlabel("MSRP (USD)")
plt.ylabel("Average Street Price (USD)")
plt.show()
```

### 💡 Expensive GPUs Get Bigger Price Markups

### *1.As MSRP goes up, the gap between MSRP and actual price also goes up — cheaper GPUs (approx 9K USD) vary by approx 7K USD, expensive ones (approx 42K USD) vary by approx 35K USD.*
### *2.Sold Out (orange) dots are mostly at the top of each group, In Stock (green) at the bottom — same "scarcity = higher price" pattern we saw in the boxplot*

```python
# Correlation between your key numeric columns:

cols = ["units_sold", "revenue_usd", "price_premium_pct", "customer_satisfaction_score", "msrp_usd"]
corr = df[cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Key Metrics",fontfamily="Times New Roman",fontsize=15)
plt.show()
```

### *💡 Price Premiums Hurt Satisfaction, MSRP Drives Revenue*

### *1.Higher MSRP GPUs bring in more revenue (0.78 correlation) but sell fewer units (-0.45) — expensive, low-volume GPUs like B200 still win on revenue.*
### *2.Paying a bigger premium over MSRP is linked to lower customer satisfaction (-0.67) — this connects back to the stock scarcity finding: chasing scarce GPUs may come at the cost of customer happiness.*

## **📝 Business Insights**

### **◾ B200 leads in revenue despite lower unit sales — Nvidia's growth comes from high-value data-center GPUs, not just volume.**
### **◾ Sold-out stock drives the highest price premiums, but higher premiums are linked to lower satisfaction — scarcity boosts revenue short-term but risks customer trust.**
### **◾ Price gaps over MSRP grow larger for expensive GPUs — enterprise/AI buyers tolerate bigger dollar markups than gaming customers.**
### **◾ Retail/Etail is the top revenue channel, and Gaming is the top customer segment.**

## **🧿 Future Recommendations**

### **• Nvidia should keep prioritizing high-margin data-center GPUs (like B200/H100) over chasing gaming GPU volume, since that's where the real revenue growth is.**
### **• Nvidia should improve supply forecasting for high-demand GPUs to reduce stockouts — protecting long-term customer trust instead of relying on scarcity-driven price spikes.**
### **• Nvidia could consider tighter pricing/allocation controls for high-end enterprise GPUs, since large unofficial markups on already-expensive chips may push serious buyers toward competitors.**
### **• Nvidia should keep investing in retail channel while also testing ways to grow the smaller channels, so revenue isn't overly dependent on just one buyer type.**
